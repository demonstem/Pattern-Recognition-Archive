**T2**:   
$$P(\omega_1 \vert x) = P(\omega_2 \vert x) \\
P(x \vert \omega_1) \cdot P(\omega_1) = P(x \vert \omega_2) \cdot P(\omega_2) \\  
\because \text{Assume}\ P(\omega_1) = P(\omega_2) \ \  \therefore P(x \vert \omega_1) = P(x \vert \omega_2) \\
$$  
$$
\dfrac{1}{2\sqrt{2\pi}} \cdot e^{-\dfrac{(x - 4) ^ 2}{2\cdot2^2}} = \dfrac{1}{2\sqrt{2\pi}} \cdot e^{-\dfrac{(x - 0) ^ 2}{2\cdot2^2}} \\
$$  
$$ 
x = 2 \\
\therefore \ \ \text{Decision boundary is }x=2
$$
